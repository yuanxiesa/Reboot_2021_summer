```python
import skimage.io
import numpy as np
import matplotlib.pyplot as plt
import skimage.draw
```

```python
# load the maize seeding picture
maize = skimage.io.imread(fname='img_processing_data/maize-roots.tif')
plt.imshow(maize)
```

```python
# create a mask of the size of the picture
mask = np.ones(shape=maize.shape[0:2], dtype='bool')
```

```python
# draw a rectangular black area in this blank mask by specifying the upper left and down right corner
rr,cc = skimage.draw.rectangle(start = (357,44), end = (740, 720))
```

```python
# see what rr and cc are
rr.shape
```

```python
cc.shape
```

```python
# rr and cc are 677 by 384 arrays
```

```python
mask[rr,cc] = False # why the assignment is done by rr,cc?
plt.imshow(mask, cmap='gray')
```
